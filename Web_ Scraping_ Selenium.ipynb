https://www.flipkart.com/apple-iphone-11-black-64-gb/productreviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART

```python
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYIJLDFV&marketplace=FLIPKART')
```

```python
ratingg = []
full_Revieww = []
Summaryy = []
```

```python
start = 0
end = 10
for page in range(start,end):
    rate = driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    for i in rate:
        ratingg.append(i.text)
    revw = driver.find_elements(By.CLASS_NAME,'_2-N8zT')
    for j in revw:
        Summaryy.append(j.text)
    full_r = driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    for k in full_r:
        full_Revieww.append(k.text)
    
    next_button = driver.find_element(By.CLASS_NAME,'_1LKTO3')
    next_button.click()
    time.sleep(3)
    
```

```python
df3 = pd.DataFrame({'ratingg':ratingg,'Summaryy':Summaryy,'full_Revieww':full_Revieww})
df3
```

### Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:

1.First get the webpagehttps://www.azquotes.com/
    
2.Click on TopQuote

3.Than scrap a)Quote b) Author c) Type Of Quotes

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.azquotes.com/')
```

```python
top = driver.find_element(By.XPATH,'//div[@class="mainmenu"]/ul/li[5]/a')
top.click()
```

```python
Qoutes = []
Author = []
type_of_Qoutes = []
for page in range(0,10):
    a = driver.find_elements(By.XPATH,'//div[@class="wrap-block"]/p/a[2]')
    for i in a:
        Qoutes.append(i.text)
    b = driver.find_elements(By.XPATH,'//div[@class="author"]/a')
    for j in b:
        Author.append(j.text)
    c = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in c:
        type_of_Qoutes.append(i.text)
        
    next_button = driver.find_element(By.XPATH,'//li[@class="next"]/a')
    next_button.click()
    time.sleep(3)
```

```python
print(len(Qoutes))
```

```python
print(len(type_of_Qoutes))
```

```python
print(len(Author))
```

```python
df = pd.DataFrame({'Qoutes':Qoutes,'Author':Author,'type_of_Qoutes':type_of_Qoutes})
df
```
